```python
#Eiqui vamas calcular a distancia entre o punto A e o punto C
#Creamos a lista x e y nde iremos metendo os desplazamentos que faga o aparato no eixe x e no eixe y.
x=[]
y=[]
n=0
#Recorremos a list
for movement in final_list_3:
    n=n+1
    #Se o movement  right, collemos o seu valor e calculamos o desplazamento en x e o desplazamento en y utilizando o seno
    #e o coseno  sabendo o grado que xira o aparato.
    if movement == 'right':
        rotating_movement=final_list_3[n]
        rotating_forward=final_list_3[n+2]
        x_mv=(math.cos(math.radians(int(rotating_movement))))*int(rotating_forward)
        y_mv=(math.sin(math.radians(int(rotating_movement))))*int(rotating_forward)
        x.append(x_mv)
        y.append(y_mv)
    #Se o movement  left, facemos o mesmo que para o right coa pequena diferencia de cambiar o signo d ngulo para indicar
    #que est indo en direccin contraria.
    elif movement == 'left':
        rotating_movement=final_list_3[n]
        rotating_forward=final_list_3[n+2]
        x_mv=(math.cos(math.radians(-int(rotating_movement))))*int(rotating_forward)
        y_mv=(math.sin(math.radians(-int(rotating_movement))))*int(rotating_forward)
        x.append(x_mv)
        y.append(y_mv)
#Coordenada en X. Facemos a suma da lista para saber o desplazamento no eixe x do aparato
coordenada_x=sum(x)
#Coordenada en Y. Facemos a suma da lista para saber o desplazamento no eixe y do aparato
coordenada_y=sum(y)
print('O desplazamento no eixe x  de %f metros e no eixe y  de %f metros ' % (coordenada_x, coordenada_y))
```

```python
#Distancia entre el punto A y el punto B (Mido a disancia en lnea recta)
#Utilizando o teorema de Pitgoras calculo a distancia entre o punto A e o punto C.
distance=math.sqrt(coordenada_x**2 + coordenada_y**2)
print('A distancia en lnea recta entre o punto A e o C  de %d metros ' % (distance))
```
