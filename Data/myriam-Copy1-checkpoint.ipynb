```python
#E eiqui comeza o jaleo. Estannos preguntando o seguinte:
#Write code in any language, that uses the data stored in Cosmos DB, 
#to find out the distance travelled in metres between markers A and C.
#Polo tanto o que fago e quedarme ca parte da lista que me interesa. E dicir, quédome dende o punto A ata a C
#e gardoo noutra lista nova a que chamo list_to_sum
start_from = 'A' # value you want to start with
finish_in='C' #value you want to finish
final_list = words_analysis[words_analysis.index(start_from):words_analysis.index(finish_in)]
print(final_list)
```

```python
n=0
suma=[]
for movement in final_list:
    n=n+1
    if movement == 'forward':
        movement_forward=final_list[n]
        movement_forward_number= int(movement_forward)
        suma.append(movement_forward_number)
resultado=sum(suma)
print('O aparato viaxa %d metros ' % (resultado))
        
```
